```python
import pandas as pd
data = pd.read_csv('Walmart (2)(1)(1)(1).csv')
data.head()
```

```python
# Question 1: Find the average of total sales for days that are holidays and days that are non-holidays.
avg_sales_holiday = data[data['Holiday_Flag'] == 1]['Weekly_Sales'].mean()
avg_sales_non_holiday = data[data['Holiday_Flag'] == 0]['Weekly_Sales'].mean()
avg_sales_holiday, avg_sales_non_holiday
```

```python
# Question 2: Provide descriptive statistics of all features in the dataset
data.describe()
```

```python
# Question 3: Complete a correlation matrix
correlation_matrix = data.corr()
correlation_matrix
```

```python
import matplotlib.pyplot as plt

# Question 4: Create a scatter plot comparing sales to each feature
features = ['Store', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']
fig, axs = plt.subplots(len(features), figsize=(5, 20))

for i, feature in enumerate(features):
    axs[i].scatter(data[feature], data['Weekly_Sales'])
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel('Weekly_Sales')
    axs[i].set_title(f'Weekly_Sales vs {feature}')

plt.tight_layout()
plt.show()
```

```python
# Question 5: Plot Weekly sales vs date
# First, convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

# Then, aggregate by date and sum the weekly sales numbers
sales_by_date = data.groupby('Date')['Weekly_Sales'].sum()

# Finally, plot the data
plt.figure(figsize=(10, 5))
plt.plot(sales_by_date)
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.title('Weekly Sales vs Date')
plt.show()
```
