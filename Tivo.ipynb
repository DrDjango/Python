# TIVO Survey Case Data

```python
import pandas as pd

# Import the dataset
tivo_data = pd.read_excel('./Tivo Survey Case Data(1)(1)(1)(1).xlsx')
tivo_data.head()
```

```python
# Categorizing each data field by its segmentation parameter type
segmentation_types = {
    'Demographic': ['Gender', 'Marital Status', 'Work Status', 'Education', 'Annual Income (x1000 $)', 'Age', 'Location'],
    'Purchasing': ['Purchasing Decision-maker', 'Purchasing Location', 'Monthly Electronics Spend', 'Monthly Household Spend', 'Purchasing Frequency (every x months)'],
    'Behavioral': ['Technology Adoption', 'TV Viewing (hours/day)'],
    'Attitude': ['Favorite feature'],
    'Other': ['ID']
}

segmentation_types

import pprint
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(segmentation_types)
```

```python
# 2a: Insert a column labeled “Annual spending on electronics”
tivo_data['Annual spending on electronics'] = tivo_data['Monthly Electronics Spend'] * 12

# 2b: Insert another column labeled “Spending as % of income”
tivo_data['Spending as % of income'] = (tivo_data['Annual spending on electronics'] / (tivo_data['Annual Income (x1000 $)'] * 1000)) * 100

tivo_data[['Annual spending on electronics', 'Spending as % of income']].head()
#tivo_data.head()
```

```python
# 3: Create a table for each attribute and record the percentage of responses for each answer
attributes = ['Gender', 'Marital Status', 'Work Status', 'Education', 'Location', 'Purchasing Decision-maker', 'Technology Adoption', 'Favorite feature']
percentage_responses = {}

for attribute in attributes:
    percentage_responses[attribute] = (tivo_data[attribute].value_counts(normalize=True) * 100).to_dict()
    print(f' % of repsondents for attribute:  {attribute}')
    pp.pprint(percentage_responses[attribute])
    print("\n\n\n")

#pp.pprint(percentage_responses)
```

```python
# 4: Create a correlation matrix for the data set
correlation_matrix = tivo_data.corr()

# Display the correlation matrix
correlation_matrix
```

```python
# Extracting the specified correlations
correlations = {
    'Age vs. Purchasing Frequency': correlation_matrix.loc['Age', 'Purchasing Frequency (every x months)'],
    'Age vs. TV Viewing': correlation_matrix.loc['Age', 'TV Viewing (hours/day)'],
    'Annual Income vs. TV Viewing': correlation_matrix.loc['Annual Income (x1000 $)', 'TV Viewing (hours/day)'],
    'Monthly Electronic Spending vs Monthly Household Spending': correlation_matrix.loc['Monthly Electronics Spend', 'Monthly Household Spend']
}




cor = pd.DataFrame(data=correlations, index = ["Correlation Coef."])
cor
```

```python
# Identifying any relationships that are high enough to make one of the attributes redundant
redundant_attributes = correlation_matrix[(correlation_matrix > 0.8) & (correlation_matrix != 1)].stack().to_dict()
redundant = pd.DataFrame(data = redundant_attributes, index =["Correlation Coef for redundancy"])
redundant
```

```python
# 5a: How many married men who are early adopters have monthly electronics spend high enough that they can afford to purchase a TiVo for $499
# and still be able to spend more on electronics in the next two years?
condition_5a = (tivo_data['Gender'] == 'male') & (tivo_data['Marital Status'] == 'married') & (tivo_data['Technology Adoption'] == 'early') & ((tivo_data['Monthly Electronics Spend'] * 24) - 499 > 0)
#tivo_data[condition_5a].shape[0]
print(f' {tivo_data[condition_5a].shape[0]} married men who are early adopters have monthly electronics spend high enough that they can afford to purchase a TiVo for $499 and still be able to spend more on electronics in the next two years ')

# 134 married men who are early adopters have monthly electronics spend high enough that they can afford to purchase a TiVo for $499
# and still be able to spend more on electronics in the next two years? 

```

```python
# 5b: How many women with an education level of MA or PhD are making purchasing decisions for electronics without discussing them with a spouse?
condition_5b = (tivo_data['Gender'] == 'female') & (tivo_data['Education'].isin(['MA', 'PhD'])) & (tivo_data['Purchasing Decision-maker'] == 'single')
print(f' {tivo_data[condition_5b].shape[0]} women with an education level of MA or PhD are making purchasing decisions for electronics without discussing them with a spouse')

```

```python
# 5c: How many early adopters purchase electronics at least once every year, and do so in stores that specialize in electronics?
condition_5c = (tivo_data['Technology Adoption'] == 'early') & (tivo_data['Purchasing Frequency (every x months)'] >= 1) & (tivo_data['Annual spending on electronics'] > 0)

print(f' {tivo_data[condition_5c].shape[0]} early adopters purchase electronics at least once every year, and do so in stores that specialize in electronics')
```

```python
# 5d: How many seniors (over the age of 65) spend more than six hours a day watching TV?
condition_5d = (tivo_data['Age'] > 65) & (tivo_data['TV Viewing (hours/day)'] > 6)
seniors_data = tivo_data[condition_5d]
count_5d = seniors_data.shape[0]
income_range_5d = (seniors_data['Annual Income (x1000 $)'].min(), seniors_data['Annual Income (x1000 $)'].max())
average_income_5d = seniors_data['Annual Income (x1000 $)'].mean()

print(f' {count_5d} early adopters purchase electronics at least once every year, and do so in stores that specialize in electronics with income range: $ {income_range_5d} and average income: $ {average_income_5d}')
```

```python
# 6: Create a groupby for each of the specified variables and show the variable and annual spend in the format of “.describe()” function.

# 6a. Gender with Monthly Electronic Spend
groupby_gender = tivo_data.groupby('Gender')['Monthly Electronics Spend'].describe()

# 6b. Martial Status with Monthly Electronic Spend
groupby_marital_status = tivo_data.groupby('Marital Status')['Monthly Electronics Spend'].describe()

# 6c. Education with Monthly Electronic Spend
groupby_education = tivo_data.groupby('Education')['Monthly Electronics Spend'].describe()

print(f' Gender with Monthly Electronic Spend:\n{groupby_gender}\n\n\nMartial Status with Monthly Electronic Spend:\n{groupby_marital_status}\n\n\nEducation with Monthly Electronic Spend:\n{groupby_education}')
```

```python
# 7: Based on the analysis of the dataset, determine the ideal customer group for marketing focus

# By considering the following factors:
#   1. Above average High Monthly Electronics Spend
#   2. Above Average Early technology adopters
#   3. Above average High TV Viewing hours (potential users for TiVo)

ideal_customer_group = tivo_data[(tivo_data['Monthly Electronics Spend'] > tivo_data['Monthly Electronics Spend'].mean()) &
                                 (tivo_data['Technology Adoption'] == 'early') &
                                 (tivo_data['TV Viewing (hours/day)'] > tivo_data['TV Viewing (hours/day)'].mean())]

# Describing the ideal customer group based on key attributes
ideal_customer_description = ideal_customer_group[['Gender', 'Marital Status', 'Education', 'Work Status', 'Location']].describe()
ideal_customer_description.iloc[2]


print(f'The ideal customer group based on key attributes : \n{ideal_customer_description.iloc[2]}')
```
